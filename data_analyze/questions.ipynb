```python
import pandas as pd

midwest = pd.read_csv('midwest.csv')
```

```python
#1
midwest.describe()
```

```python
#변수명 바꾸기
midwest = midwest.rename(columns={'poptotal':'total'})
midwest = midwest.rename(columns={'popasian':'asian'})
```

```python
midwest.head()
```

```python
#백분율 파생변수
midwest['ratio'] = (midwest['asian']/midwest['total']) * 100 
```

```python
midwest.head()
```

```python
#백분율 히스토그
midwest['ratio'].plot.hist()
```

```python
#평균초과시 large, 그외엔 small
import numpy as np
avg = midwest['asian'].mean()
midwest['avg'] = np.where(midwest['asian'] > avg, 'large','small')
```

```python
midwest.head()
```

```python
#5 large, small 빈도 그래프
midwest['avg'].value_counts().plot.bar()
```
